Task 1

Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable,such as the distribution of ages or genders in population

Data set used - population_by_age_group from kaggle

Importing and previewing the data

```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#reading the data from csv file
file_path="population_by_age_group.csv"
data=pd.read_csv(file_path)

#previewing the first 5 records
print(data.head(5))

#previewing the last 5 records
print(data.tail(5))
                 
```

```python
#Getting an overview of the data set
print('Shape:',data.shape)

#retrieving column names
print('Columns:',data.columns)

#searching for dulicate rows
print("duplicated rows:",data.duplicated().sum())

#Checking for missing data
print("missing data :")
print(data.isna().sum())
```

Finding and plotting the top 10 countries with the highest population


```python
#sorting the data by population
sorted_data=data.sort_values(by='Total',ascending=False)

#finding the top 10 countries
top_10_countries=sorted_data.head(10)
print(top_10_countries[['Country','Total']])

top_10_country_name=top_10_countries['Country']
top_10_country_pop=top_10_countries['Total']
```

```python
#create the bar plot
plt.figure(figsize=(15,8))
sns.barplot(data=top_10_countries,y=top_10_country_name,x=top_10_country_pop,palette='viridis')

#add titles and labes
plt.title("Top 10 Countries by Total Population",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Countries",fontsize=10)

#rotate axis for better readability
plt.yticks(rotation=45)
plt.xticks(rotation=45)

#adjust layout
plt.tight_layout()
#show the plot
plt.show()
```

Finding and plotting the 10 least populated countries

```python
#sorting the data by population
sorted_data_asc=data.sort_values(by='Total',ascending=True)

#finding the top 10 countries
bot_10_countries=sorted_data_asc.head(10)
print(bot_10_countries[['Country','Total']])

bot_10_country_name=bot_10_countries['Country']
bot_10_country_pop=bot_10_countries['Total']
```

```python
#create the bar plot
plt.figure(figsize=(15,8))
sns.barplot(data=bot_10_countries,y=bot_10_country_name,x=bot_10_country_pop,palette='viridis')

#set titles and labels
plt.title("10 Countries with the least population",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Countries",fontsize=10)

#rotate axis for better readability
plt.yticks(rotation=45)
plt.xticks(rotation=45)

#adjust layout
plt.tight_layout()

#show the plot
plt.show()
```

Filtering data for a particular country

```python
#extract data by country name
India_data=pd.DataFrame(data[data['Country']=='India'])

#remove unnecessary columns
India_data=India_data.drop(columns=['Country','Total'])
print(India_data)
```

Creating the bar plot for Age_range vs Population count

```python

#create the bar plot
plt.figure(figsize=(15,8))
sns.barplot(data=India_data,palette='viridis')

#set titles and labels
plt.title("India Population",fontsize=10)
plt.xlabel('Age Range',fontsize=10)
plt.ylabel('Population',fontsize=10)

# Rotate axis labels for better readability
plt.xticks(rotation=45)
plt.yticks(rotation=45)

#adjust the layout
plt.tight_layout()

#show the plot
plt.show()
```
